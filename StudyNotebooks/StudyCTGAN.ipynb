```python
import os
os.environ["OMP_NUM_THREADS"] = "1"

from ctgan import CTGAN, load_demo

real_data = load_demo()

# Names of the columns that are discrete
discrete_columns = [
    'workclass',
    'education',
    'marital-status',
    'occupation',
    'relationship',
    'race',
    'sex',
    'native-country',
    'income'
]

ctgan = CTGAN(epochs=5, 
              embedding_dim=64, 
              generator_dim=(64, 64), 
              discriminator_dim=(64, 64), 
              batch_size=100)
ctgan.fit(real_data.iloc[:2000], discrete_columns)

# Create synthetic data
synthetic_data = ctgan.sample(100)
```
